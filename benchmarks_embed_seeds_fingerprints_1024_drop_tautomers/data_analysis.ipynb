# Let's make the table with mean and std for seeds from 0 to 9

```python
path_dict = {
    'LabelEncoder': 'sklearn_LabelEncoder_seed_*.csv',
    'OneHotEncoder': 'sklearn_OneHotEncoder_seed_*.csv',
    'CategoryEmbedding': 'sklearn_EmbeddingCategoryEmbedding_seed_*.csv',
    'AutoInt': 'sklearn_EmbeddingAutoInt_seed_*.csv',
    'TabTransformer': 'sklearn_EmbeddingTabTransformer_seed_*.csv',
}

mean_df, std_df = pd.DataFrame(), pd.DataFrame()
dummy_df = pd.read_csv('sklearn_LabelEncoder_seed_42.csv')
mean_df['Model'] = dummy_df['Model']
std_df['Model'] = dummy_df['Model']
for key, value in path_dict.items():
    seeds_path = sorted(glob.glob(value))
    lb_wpc_arr = np.zeros((len(seeds_path), len(dummy_df)))
    for i, path in enumerate(seeds_path):
        df = pd.read_csv(path)
        lb_wpc_arr[i] = df['LB WPC']
    mean_df[key] = lb_wpc_arr.mean(axis=0)
    std_df[key] = lb_wpc_arr.std(axis=0)
mean_df
    
```

```python
std_df
```

```python
import pandas as pd
import numpy as np

# Your mean_df and std_df must be defined already

# Define encoders
traditional = ['LabelEncoder', 'OneHotEncoder']
neural = ['CategoryEmbedding', 'AutoInt', 'TabTransformer']
all_models = traditional + neural

# Store final LaTeX lines
latex_lines = []

# Header
latex_lines.append(r'\footnotesize')
latex_lines.append(r'\begin{tabular}{lccccc}')
latex_lines.append(r'\hline')
latex_lines.append(r'                 & \multicolumn{2}{c}{\textbf{Traditional encodings}}  & \multicolumn{3}{c}{\textbf{Neural Network encodings}}                             \\ \hline')
latex_lines.append(r'\textbf{Model}    & \textbf{LabelEncoder}    & \textbf{OneHotEncoder}   & \textbf{CategoryEmbedding} & \textbf{AutoInt}          & \textbf{TabTransformer}  \\ \hline')

# Process each row
for i in range(len(mean_df)):
    model_name = mean_df.loc[i, 'Model']
    values = {}
    best_val = -np.inf
    best_key = None

    for key in all_models:
        mean = mean_df.loc[i, key]
        std = std_df.loc[i, key]
        val_str = f"{mean:.3f}"
        if std > 0:
            val_str += f" ({std:.3f})"
        values[key] = val_str

        if mean > best_val:
            best_val = mean
            best_key = key

    # Check if asterisk condition is met (only for neural encoders)
    asterisk = ""
    if best_key in neural:
        label_val = mean_df.loc[i, 'LabelEncoder']
        onehot_val = mean_df.loc[i, 'OneHotEncoder']
        if (best_val >= 1.3 * label_val) and (best_val >= 1.3 * onehot_val):
            asterisk = "*"

    # Replace best value with bold
    values[best_key] = r'\textbf{' + values[best_key] + asterisk + r'}'

    row = f"{model_name} & " + " & ".join(values[key] for key in traditional + neural) + r" \\"
    latex_lines.append(row)

latex_lines.append(r'\hline')
latex_lines.append(r'\end{tabular}')
latex_lines.append(r'\caption*{\small An asterisk (*) indicates that the embedding method achieves at least 30\% better performance compared to both Label Encoding and OneHot Encoding.}')
latex_lines.append(r'\label{table:encodinbenchmarks_embed_seeds_fingerprints_1024/data_analysis.ipynbg_comparison}')

# Output
latex_code = "\n".join(latex_lines)
print(latex_code)
```
